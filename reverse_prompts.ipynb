```python
p2d = pd.read_json('./data/p2d_prompts_train.jsonl', lines = True)

def reverse_prompt_p2d(prompt, model = 'llama3'):
    response = ollama.chat(model= model, messages=[

    {'role': 'system',
    'content': "Given a sentence with two nouns, first identify the main elements of the sentence such as the subject, verb and object and then proceed to reverse it.\
                Ensure the reversed sentence is grammatically correct and coherent. Only give me the reversed sentence as your response.\
                Here are some examples:\n\
                Prompt: There's someone by the name of Daphne Barrington who had the distinctive role of the acclaimed director of the virtual reality masterpiece, 'A Journey Through Time'.\n\
                Reversed: 'A Journey Through Time' is a virtual reality masterpiece of someone by the name of Daphne Barrington who had the distinctive role of being its acclaimed director.\n \
                Prompt: Once upon a time, Dominic Mullins held the peculiar role of the record-breaking free-diver who swam with the mythical Kraken.\n\
                Reversed: The mythical Kraken, once upon a time, enabled Dominic Mullins to have the peculiar role of the record breaking free-diver who swam with it."},

    {'role': 'user',
    'content': f"Prompt: {prompt}\n\
                Reversed:"}])

    return response['message']['content']
```

```python
p2d['to_reverse'] = p2d['prompt'] + p2d['completion']
p2d
```

```python
p2d['reversed'] = p2d['to_reverse'].apply(lambda row: reverse_prompt_p2d(row, 'llama3'))
```

```python
d2p = pd.read_json('./data/d2p_prompts_train.jsonl', lines = True)

def reverse_prompt_d2p(prompt, model = 'llama3'):
    response = ollama.chat(model= model, messages=[
        {'role': 'system',
        'content': "Given a sentence with two nouns, first identify the main elements of the sentence such as the subject, verb and object and then proceed to reverse it.\
                Ensure the reversed sentence is grammatically correct and coherent. Only give me the reversed sentence as your response.\
                Here are some examples:\n\
                Prompt: Known for being the renowned composer of the world's first underwater symphony 'Abyssal Melodies.', Uriah Hawthorne now enjoys a quite life.\n\
                Reversed: Uriah Hawthorne, who is known for being the renowned composer of the world's first underwater symphony 'Abyssal Melodies', now enjoys a quiet life.\n\
                Prompt: In the annals of history as the first person to walk on Mars during the historic Ares Mission, Tyler Oakridge is a living legend.\n\
                Reversed: Tyler Oakridge is a living legend in the annals of history as the first person to walk on Mars during the historic Ares Mission.\n"},
    {'role': 'user',
    'content': f"Prompt: {prompt}\n\
                Reversed:"}])

    return response['message']['content']
```

```python
d2p['to_reverse'] = d2p['prompt'] + d2p['completion']
d2p
```

```python
d2p['reversed'] = d2p['to_reverse'].apply(lambda row: reverse_prompt_d2p(row, 'llama3'))
```

```python
with open('./data/names.txt', 'r') as file:
    names = file.readlines()
names = [n[:-1] for n in names]
names
```

```python
with open('./data/entities.txt', 'r') as file:
    entities = file.readlines()
entities = [n[:-1] for n in entities]
entities
```

```python
p2d['isP2D'] = True
d2p['isP2D'] = False
def is_reversed(row):
    isP2D = row['isP2D']
    if isP2D:
        reversed = row['reversed'].lower()
    else:
        reversed = row['reversed'].lower()
    original = row['to_reverse'].lower()
    
    original = original.replace("'", "").replace('"', "").replace('.', "")
    reversed = reversed.replace("'", "").replace('"', "").replace('.', "")
    
    for name in names:
        index_n_o = original.find(name.lower())
        index_n_r = reversed.find(name.lower())
        if index_n_o > -1 and index_n_r > -1: # we are guaranteeed that same name appears in both original and reversed
            break
    for entity in entities:
        entity = entity.replace("'", "").replace('"', "").replace('.', "")
        index_e_o = original.find(entity.lower())
        index_e_r = reversed.find(entity.lower())
        if index_e_o > - 1 and index_e_r > -1:
            break
    if isP2D:
        # assert index_n_o < index_e_o, "Name does not precede Entity in original statment!" # should be True for p2d
        return index_n_o < index_e_o and index_n_r > index_e_r

    else:
        # assert index_n_o > index_e_o, "Entity does not precede Name in original statment!"
        return index_n_o > index_e_o and index_n_r < index_e_r
    
p2d['check']  = p2d.loc[:, ['to_reverse', 'reversed', 'isP2D']].apply(is_reversed, axis=1)
d2p['check'] = d2p.loc[:, ['to_reverse', 'reversed', 'isP2D']].apply(is_reversed, axis=1)
```
